```python
def find_boundary(data, variables):
    Q1 = data[variables].quantile(.25)
    Q3 = data[variables].quantile(.75)
    IQR = Q3 - Q1
    lower_boundary = Q1 - 1.5*IQR
    upper_boundary = Q3 + 1.5*IQR
    
    return lower_boundary, upper_boundary
```

```python
lower_fixed, upper_fixed = find_boundary(df, 'Blast Furnace Slag (component 2)(kg in a m^3 mixture)')
print("upper_fixed for Blast_Furnace is: ", upper_fixed)
print("lower_fixed for Blast_Furnace is: ", lower_fixed)
```

```python
df['Blast Furnace Slag (component 2)(kg in a m^3 mixture)'] = np.where(df['Blast Furnace Slag (component 2)(kg in a m^3 mixture)']>upper_fixed, upper_fixed, 
                                np.where(df['Blast Furnace Slag (component 2)(kg in a m^3 mixture)']<lower_fixed, lower_fixed, df['Blast Furnace Slag (component 2)(kg in a m^3 mixture)']))
```
